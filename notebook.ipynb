## Visualization & Explanatory Analysis

### Pertanyaan 1:

Produk manakah yang paling sedikit dan paling banyak dijual?

```python
product_id_counts = all_data.groupby('product_category_name')['product_id'].count().reset_index()
sorted_df = product_id_counts.sort_values(by='product_id', ascending=False)

sorted_df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#ffc0cb", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="product_id", y="product_category_name", data=sorted_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Produk Paling Banyak Terjual", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="product_id", y="product_category_name", data=sorted_df.sort_values(by="product_id", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk Paling Sedikit Terjual", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Produk Paling Banyak dan Sedikit Terjual", fontsize=20)
plt.show()
```

Berdasarkan diagram di atas, produk yang paling banyak terjual merupakan casa_mesa_banho, sedangkan produk yang paling sedikit terjual adalah seguros e servicos.

### Pertanyaan 2:

Berapa banyak pengeluaran yang dikeluarkan customer dalam beberapa bulan terakhir?

```python
monthly_spend_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "payment_value":"sum"
})
monthly_spend_df.index = monthly_spend_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly_spend_df = monthly_spend_df.reset_index()
monthly_spend_df.rename(columns={
    "payment_value":"total_spend"
}, inplace=True)
monthly_spend_df.head()
```

```python
monthly_spend_df = monthly_spend_df.sort_values('total_spend').drop_duplicates('order_approved_at', keep='last')
```

```python
monthly_spend_df.head()
```

```python
monthly_spend_df.sort_values(by='total_spend')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_spend_df["order_approved_at"],
    monthly_spend_df["total_spend"],
    marker='o',
    linewidth=2,
    color="#FFC0CB"
)
plt.title("Total Pengeluaran Kustomer per bulan (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan grafik di atas, uang yang paling dikeluarkan adalah pada bulan November dan paling sedikit bulan September.

### Pertanyaan 3:

Bagaimana performa E-Commerce selama selang waktu terakhir?

```python
monthly_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_df.index = monthly_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
monthly_df.head()
```

```python
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
```

```python
monthly_df.head()
```

```python
monthly_df.sort_values(by='order_count')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_df["month_numeric"] = monthly_df["order_approved_at"].map(month_mapping)
monthly_df = monthly_df.sort_values("month_numeric")
monthly_df = monthly_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_df["order_approved_at"],
    monthly_df["order_count"],
    marker='o',
    linewidth=2,
    color="#ffc0cb"
)
plt.title("Angka Penjualan Per Bulan (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan grafik di atas, order tertinggi berada di bulan November, sedangkan terendah di bulan September.

### Pertanyaan 4:

Bagaimana tingkat kepuasan customer?

```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))
sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            palette=["#ffc0cb" if score == most_common_score else "#D3D3D3" for score in review_scores.index]
            )

plt.title("Rating Kustomer untuk Pelayanan", fontsize=15)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(fontsize=12)
plt.show()
```

Berdasarkan diagram di atas menunjukkan bahwa pelanggan memiliki kepuasan yang tinggi, dibuktikan dengan bintang 5 memiliki data tertinggi dibandingkan data rating lainnya.
